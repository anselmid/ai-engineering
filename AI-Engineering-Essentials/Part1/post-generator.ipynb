### Step 1: Import Libraries and API Keys

```python
# Import the libraries
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("API key is missing")
```

### Step 2: Searching the web

```python
# Now search the web for the latest AI conferences in Dubai. Anthropic also has a model that can search the web.
client = OpenAI()

# You cant use temperature with this model
completion = client.chat.completions.create(
    model="gpt-4o-mini-search-preview",     # This is a specical model that you can use to search the web. You can't chat with it.
    messages=[
        {"role": "user", "content": "What are the latest AI confrerences happening in Dubai?"}
    ],
    max_tokens=500
)
```

```python
# and view the results
search_results = completion.choices[0].message.content
display(Markdown(search_results))
```

### Step 3: Summarize and write a social media post

```python
# Finally summarize the results and create a social media post. One of the issues with LLMs is that they can only respond to what they have training on.
# So if you want them to respond to something new, you have to provide the context in the prompt.
# Here we provide the context of the search results to the model and ask it to summarize and create a social media post.

system_message = """
You are a helpful assistant whos job is to accurately summarize long series of 
news articles provided to you by the user and write up a compelling and easy to 
digest social media post.
"""

prompt = f"""
The following is a series of news articles on the field of AI. Please summarize
the articles and write a compelling social media post for LinkedIn highlighting
key points about each article. 

The post should be short and concise and should not be more than 100 words.

{search_results}

"""

completion = client.chat.completions.create(
    model="gpt-4o-mini",    
    messages=[
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ],
    max_tokens=500
)

chat_response = completion.choices[0].message.content
display(Markdown(chat_response))
```
